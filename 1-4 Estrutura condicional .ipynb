```python
# if = "Si" acontecer algo
if 5 > 10:
  print('Verdade') # se o que foi pedido for falso ele nao mostrara nada.
```

```python
# if/else  # else = "E si caso?"
num1 = 10
num2 = 15
if num1 > num2:
  print('O num1 é o maior número')
else:
  print('O num2 e o maior número')
```

```python
#if/ elif / else   #elif (elif é uma abreviação de "else if")= "É se?" "e se caso?"
num1 = 616
num2 = 616
if num1 > num2:
  print('O num1 é o maior numero')
elif num1 < num2:
  print('O num2 é o maior numero')
else:
  print('Os dois são iguais')
```

```python
num1 = 4
num2 = 8
num3 = 12
if (num1 > num2) and (num1 > num3):
  print('O num1 é o maior numero')
else:
  print('O num1 não é o maior numero')
```
